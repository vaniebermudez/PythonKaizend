```python
import random
import requests

from IPython import embed
from time import sleep
from bs4 import BeautifulSoup

BASE_URL = "http://sample-target-bucket-with-html-pages.s3-website-ap-southeast-1.amazonaws.com"

PAGES = ["group1/1.html",
         "group1/2.html",
         "group1/3.html",
         "group1/4.html",
         "group1/5.html",
        ]


def delay(seconds):
    print(f"Sleeping for {seconds} second(s)")
    sleep(seconds)
    
def get_random_number():
    return random.randint(1,3)

def extract_html_content(target_url):
    html_string = requests.get(target_url)
    return html_string.text

def extract_target_value_from_page(html_string):
    html_doc = html_string
    soup = BeautifulSoup(html_doc, 'html.parser')
    target = soup.find('div').get_text()
    return target

def main():
    for page in PAGES:
        target_page = BASE_URL + "/" + page
        extracted_html = extract_html_content(target_page)
        page_value = extract_target_value_from_page(extracted_html)
        print(target_page)
        print(extracted_html)
        print(page_value + "\n")
        
        delay(get_random_number())
        
if __name__ == "__main__":
    main()
```
