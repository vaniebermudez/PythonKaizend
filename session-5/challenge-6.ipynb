```python
import random
import requests

from IPython import embed
from time import sleep
from bs4 import BeautifulSoup

BASE_URL = "https://sample-target-bucket-with-html-pages.s3-ap-southeast-1.amazonaws.com/group2/index.html"


def debug_input_output(function):
    
    def wrapper(*args, **kwargs):
        print(f"[Start: {function.__name__}]")
        output = function(*args, **kwargs)
        print(f"[End: {function.__name__}]")
        
        return output
    return wrapper

@debug_input_output
def delay(seconds):
    print(f"Sleeping for {seconds} second(s)")
    sleep(seconds)
    
def get_random_number():
    return random.randint(1,3)

@debug_input_output
def extract_html_content(target_url):
    print(f"Downloading HTML content of {target_url}")
    response = requests.get(target_url)
    return response.text

@debug_input_output
def extract_target_value_from_page(html_string):
    soup = BeautifulSoup(html_string, 'html.parser')
    div_elements = soup.find_all("div", {"id": "target"})
    return div_elements
   
def initial_page(BASE_URL):
    initial_page = extract_html_content(BASE_URL)
    print(initial_page)
    soup = BeautifulSoup(initial_page, 'html.parser')
    a_elements = soup.find_all('a')
    return a_elements
    
def main():
    
    PAGES = initial_page(BASE_URL)
    print(PAGES, '\n')
    delay(get_random_number())
    
    for page in PAGES:
        target_page = BASE_URL[:-18] +page.get('href')
        print(target_page)
        
        html_content = extract_html_content(target_page)
        print(html_content)
        
        target_values = extract_target_value_from_page(html_content)
        for target_value in target_values:
            print(target_value.get_text())
            
        delay(get_random_number())
        print()

if __name__ == "__main__":
    main()
```
